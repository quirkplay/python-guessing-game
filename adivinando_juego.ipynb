```python
print("Cargando módulos...")
from configuracion import cargar_configuracion
from adivinando_modulo import (
    mostrar_titulo,
    modo_solitario,
    modo_dos_jugadores,
    menu_como_jugar,
    menu_configuracion,
    menu_estadisticas
)
print("\nMódulos cargados correctamente.\n")

# ===== PROGRAMA PRINCIPAL =====

mostrar_titulo()

config = cargar_configuracion()

salir = False

while salir == False:
    print("\n" + "="*60)
    print(" "*15 + "ADIVINA EL NÚMERO")
    print("="*60)
    print("1. Partida modo solitario")
    print("2. Partida 2 Jugadores")
    print("3. Estadísticas")
    print("4. ¿Cómo jugar?")
    print("5. Configuración")
    print("6. Salir")
    print("="*60)
    
    opcion = input("\nElige una opción: ")
    
    if opcion.isdigit():
        opcion = int(opcion)
        
        if 1 <= opcion <= 6:
            if opcion == 1:
                resultado = modo_solitario(config)
                if resultado == "salir":
                    salir = True
                
            elif opcion == 2:
                resultado = modo_dos_jugadores(config)
                if resultado == "salir":
                    salir = True
                
            elif opcion == 3:
                menu_estadisticas()
                
            elif opcion == 4:
                menu_como_jugar()
                
            elif opcion == 5:
                menu_configuracion(config)
                
            elif opcion == 6:
                print("\n¡Gracias por jugar!")
                salir = True
        else:
            print("ERROR: Debes elegir entre 1 y 6")
    else:
        print("ERROR: Debes introducir un número")

print("\n🎮 Juego finalizado. ¡Hasta la próxima!")
```
